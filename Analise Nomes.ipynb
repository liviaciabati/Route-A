```python
path = 'C:\\Users\\livia\\Dropbox\\HRP Alliance authorship paper\\Data 2022-04-14\\'
authors = pd.read_csv(path + 'authors_gender_loc.csv')
authors = authors.drop("Unnamed: 0", axis = 1)
authors = authors.drop("Unnamed: 0.1", axis = 1)
n_authors_entry = len(authors)
print('Num autores total: {}'.format(n_authors_entry))

for nome in authors.deduplicated_name_std.value_counts().index:
    if ('World Health Organization' in nome) or ('WHO' in nome.upper()) \
        or ('Research'.upper() in nome.upper()) or ('Reproductive'.upper() in nome.upper()) \
        or ('Study'.upper() in nome.upper()) or ('Health'.upper() in nome.upper()) \
        or ('GROUP' in nome.upper()) or ('NETWORK' in nome.upper()):
        authors = authors[authors.deduplicated_name_std != nome]
authors

n_authors_without_inst = len(authors)
print('N instituicoes: {}'.format(n_authors_entry - n_authors_without_inst))
print('Num autores total sem instituicoes: {}'.format(n_authors_without_inst))
```
